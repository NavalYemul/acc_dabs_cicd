```python
from pyspark.sql.functions import *
```

```python
dbutils.widgets.text("catalog","1_dev")
catalog=dbutils.widgets.get("catalog")
```

```python
df=spark.table(f"{catalog}.naval_bronze.sales")

df1=df.dropDuplicates().dropna().drop("ingestion_date","path")

df1.write.mode("overwrite").saveAsTable(f"{catalog}.naval_silver.sales_cleaned")
```

```python
df=spark.table(f"{catalog}.naval_silver.sales_cleaned")

df1=df.groupBy("product_id").agg(sum("total_amount").alias("total_sales"))

df1.write.mode("overwrite").saveAsTable(f"{catalog}.naval_gold.total_sales_product")
```
