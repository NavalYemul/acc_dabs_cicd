```python
from pyspark.sql.functions import *
```

```python
dbutils.widgets.text("catalog","1_dev")
```

```python
catalog=dbutils.widgets.get("catalog")
```

```python
df=spark.read.csv(f"/Volumes/{catalog}/naval_bronze/raw/sales.csv",header=True,inferSchema=True)
```

```python
df1=df.withColumn("ingestion_date",current_timestamp()).withColumn("path",col("_metadata.file_path"))
```
